```python
import numpy as np
import pandas as pd
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

----------------------

```python
applist = pd.read_csv('../solution3/Data/app_cnt550.csv')
brand = pd.read_csv('data/device_brand_v4.csv')
```

```python
applist.shape,brand.shape
```

```python
train_data = pd.read_csv('../Demo/deviceid_train.tsv',sep='\t',names=['device_id','sex','age'])
test_data = pd.read_csv('../Demo/deviceid_test.tsv',sep='\t',names=['device_id'])

train_data['label'] = train_data.sex.astype(str)+'-'+train_data.age.astype(str)

labelencode_dict = {}
labels = ['1-0', '1-1', '1-2', '1-3', '1-4', '1-5', '1-6', '1-7','1-8', '1-9', '1-10',
          '2-0', '2-1', '2-2', '2-3','2-4', '2-5', '2-6', '2-7', '2-8', '2-9', '2-10']
for i,label in enumerate(labels):
    labelencode_dict[label] = i
#--------------------------
train_data.label.replace(labelencode_dict,inplace=True)
```

------------------------------------------
# Merge Data

```python
#brand.drop('label',axis=1,inplace=True)
```

```python
train_data = train_data.merge(applist, on='device_id', how='left')
train_data = train_data.merge(brand,on='device_id',how='left')

test_data = test_data.merge(applist, on='device_id', how='left')
test_data = test_data.merge(brand, on='device_id', how='left')
```

```python
test_data.info()
```

```python
test_data.fillna(0,inplace=True)
```

```python
train_data.fillna(0,inplace=True)
```

```python
label = train_data[['label']]
data = train_data.drop(['device_id','sex','age','label','app_names'],axis=1)

x_train,x_valid, y_train, y_valid = train_test_split(data, label,test_size=0.2)
```

```python
xgbc = xgb.XGBClassifier(max_depth=3, learning_rate=0.10123, n_estimators=1000,
                         silent=True, objective='multi:softprob',
                         booster='gbtree', n_jobs=-1,
                         gamma=0, subsample=1,
                         colsample_bytree=1, colsample_bylevel=1.,
                         reg_alpha=0, reg_lambda=1,
                         scale_pos_weight=1,
                         base_score=0.5,
                         max_delta_step = 0,
                         random_state=666)
```

```python
xgbc.fit(x_train, y_train,
         eval_set=[(x_train, y_train),(x_valid, y_valid)],
         eval_metric = 'mlogloss',
         early_stopping_rounds=10,
         verbose=True)
```

# Feature Importance
--------------------------------------------

```python
feature_importance = pd.DataFrame()
feature_importance['feature'] = x_train.columns.values
feature_importance['importrance'] = xgbc.feature_importances_

feature_importance.set_index('feature').plot(kind='barh',figsize=(8,100))
```

# Learning Curve
---------------------------------------

```python
eval_result = xgbc.evals_result()
plt.figure(figsize=(10,5))
plt.plot(range(1,41),eval_result['validation_0']['mlogloss'],'r', label='Train logloss')
plt.plot(range(1,41),eval_result['validation_1']['mlogloss'],'b', label='Vlaid logloss')
plt.legend()
#plt.ylim(2.6,3.0)
plt.grid(True)
```

# Make submission
---------------------------------------------

```python
def make_submit(result,test_id):
    result = pd.DataFrame(result,
                          columns=['1-0', '1-1', '1-2', '1-3', '1-4', '1-5', '1-6', '1-7',
                                   '1-8', '1-9', '1-10', '2-0', '2-1', '2-2', '2-3','2-4', '2-5', '2-6', '2-7', '2-8', 
                                   '2-9', '2-10'])
    result['DeviceID'] = test_id.values
    result = result[['DeviceID']+list(result.columns[:-1])]
    return result
```

```python
test = test_data.drop(['device_id','app_names'],axis=1)
```

```python
test.shape
```

```python
result = xgbc.predict_proba(test)
```

```python
result = make_submit(result, test_data.device_id)
```

```python
result.to_csv('submit/baseline4_3.csv',index=False)
```
